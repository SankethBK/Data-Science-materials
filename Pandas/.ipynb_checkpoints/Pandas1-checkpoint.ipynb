### More about Data Frames

 We can think of dataframes as dictionaries with column names as keys and and a list of values of that column.

```python
people = {
    "first" : ["Corey" , "John", "Jane"], 
    "last" : ["Schafer", "Doe", "Doe"], 
    "email" : ["CoreySccafer@gmail.com", "JohnDoe@gmail.com", "JaneDoe@gmail.com"]
}


# We can create a dataframe by directly passing the dictionary

people_df = pd.DataFrame(people)  # note that we are not using read_csv method here 
people_df
```

```python
# Inorder to retrive a particular column 

people_df['email']  # people_df.email is also valid 
```

```python
# It returns Series type 
# We can think of each column a sa Series object and a dataframe is a container for such Series objects

type(people_df['email'])
```

```python
# Accessing multiple columns at the same time 

people_df[['last', 'email']]

# The returned object is another DataFrame 
```

```python
# Inorder to get all the column names 

people_df.columns   # Object type means string
```

```python
# To access rows we use loc and iloc (location and integer location)

people_df.iloc[0]

# It returns a Series object, but we can see that when we accessed columns indexes were integers , when we access rows ,indexes are column names
```

```python
# To access multiple rows 

people_df.iloc[[0,1]]
```

```python
# We can access columns along with rows using iloc, but since it is integer based we have to pass the indexes of columns 

people_df.iloc[[0,1], 2]  # Get 3 rd column along with first 2 rows 
people_df.iloc[[0,1], [1,2]]  # If you want multiple columns pass a list for second argument also 

```

```python
# Accessing data with loc 

# In our current example we dont have indexes for rows so it would be integers only ( similar to iloc) but with columns 
# we can pass column labels 

people_df.loc[0]  # FIrst row 
people_df.loc[[0,1]] # First 2 rows
people_df.loc[[0,1] , ["last", "email"]]  # Accessing rows and ciolumns with labels
```

```python
# Back to Stackoverflow dataset 

df['Hobbyist'] # Access the Hobbyiest column
df['Hobbyist'].value_counts()   # Getting the counts for each entry in the row
```

```python
df.loc[0, "Hobbyist"]  # To get first response for Hobbyist column 
df.loc[[0,1,2], "Hobbyist"] # To get first 3 responses for Hobbyist column
df.loc[0:10, "Hobbyist"]  # To get 0-10 row's entries for Hobbyist column here last index is inclusive 
df.loc[0:10, "Hobbyist" : "Employment"] # It will get all the columns from HobbyList to Employment, remember columns are ordered 
```

### Filtering data from Dataframes and Series objects

First we will work with people_df dataframe

```python
# Get all rows where last name = "Doe"

people_df["last"] == "Doe" # Works  , but returns a series object of only "last" column with True/False values which satisfoed the filter 

# So this acts as a mask 

filt = (people_df["last"] == "Doe")
people_df[filt]
```

```python
people_df[[True,False,True]]  # may be works like this, but definetly more complicated things are going on inside
```

```python
# Adavntages of using loc 

people_df.loc[filt, 'email']  # Along with applying filter , we can still select columns that we wan
```

```python
# Appling booleans in filters 
# & - And 
# | - Or 
# ~ - Applies negation on filter 

# & - And

filt = (people_df["last"] == "Doe") & (people_df["first"] == "John")
people_df.loc[filt, "email"]
```

```python
# \ - Or

filt = (people_df["last"] == "Schafer") | (people_df["first"] == "John")
people_df.loc[filt, "email"]
```

```python
# ^ - Negate 


filt = (people_df["last"] == "Schafer") | (people_df["first"] == "John")
people_df.loc[~filt, "email"]
```

### Updating rows and columns 


```python
# Lets consider people_df dataframe first 
# Let's say we want to rename first and last to first_name and last_name 

people_df.columns = ["first_name", "last_name", "email"]
people_df.columns
```

```python
# If you have many columns and you want to change the name of columns to upper case or change the spaces in column heading to underscore, ect we can use list comprehension

people_df.columns = [x.upper() for x in people_df.columns]
people_df.columns
```

```python
# Remove spaces and replace with underscores - inorder to use dot operator to access columns 

people_df.columns = people_df.columns.str.replace(' ', '_')

# settig back to lowercase 

people_df.columns = [x.lower() for x in people_df.columns]
```

```python
# To rename a particular column 

people_df.rename(columns = {"first_name" : "first", "last_name" : "last"}, inplace = True)
people_df
```

```python
# Updating rows 

people_df.loc[1] = ["JohnSmith@gmail.com", "John", "Smith" ]   # It is OK to use any method to grab rows
people_df
```

```python
# Inorder to change only some columns 

people_df.loc[1, ["last", "email"]] = ["Doe", "JohnDoe@gmail.com"]
people_df
```

```python
# Updating a single cell

people_df.loc[1, "email"] = "JohnDOe12@gmail.com"
people_df

# But there is an "at" method which works similar to locc may have better performance in this operation

people_df.at[1, "email"] = "JohnDoe@gmail.com"
people_df.loc[1] = ["JohnDoe@gmail.com", "John", "Doe"]
people_df.columns = ["email", "first", "last"]
people_df
```

```python
# One posssibe mistake is not to use .loc or .at for assignment of rows 

filt = (people_df["email"] == "JohnDoe@gmail.com")
people_df[filt]["last"] = "Smith"
```

```python
# Inorder to change multiple rows at once 

people_df["email"] = people_df["email"].str.lower()
people_df
```

### But inorder to do complex changes , we have four mathods 

* apply
* map 
* applymap 
* replace

```python
# apply - used for calling function on values 
# apply can work on either a dataframe or a series object

# apply - on Series object 

people_df["email"].apply(len)

def update_email(email):
    return email.upper()

people_df["email"] = people_df["email"].apply(update_email)
people_df["email"] = people_df["email"].apply(lambda x : x.lower())
people_df


# apply on dataframe 

people_df.apply(len)  # It applies len function on every series especially columns (axis = "rows")

# to apply on rows 

people_df.apply(len, axis = "columns")
```

```python
# Series have a min method we can use from apply 

people_df.apply(pd.Series.min)   
people_df.apply(lambda x : x.min())  # because series have a min() method
```

```python
# applymap - works only on dataframe
# It is applied on each object of dataframe

people_df.applymap(len)
people_df.applymap(str.lower)  # all strings to be lower
```

```python
# map - works only on a series 
# Used to substitute values in a series 
# Replaces unsubstituted values with NaN

people_df["first"].map({"Corey" : "Chris", "Jane" : "Mary"})
people_df     # original df is unchanged
```

```python
# Replace - Used to substitute values in a series 
# Keeps unsubstituted values as it is

people_df["first"].replace({"Corey" : "Chris", "Jane" : "Mary"})
```

### Handling missing data

```python
import numpy as np

people = {
    "first" : ["Corey" , "John", "Jane", "Adam", np.nan, None, "NA"], 
    "last" : ["Schafer", "Doe", "Doe", "Doe","NA",  np.nan, "Missing"], 
    "email" : ["CoreySccafer@gmail.com", "JohnDoe@gmail.com", "JaneDoe@gmail.com", "A@gmail.com", np.nan, "EMail.com", "Missing"],
    "age" : ['33', '55', '63', '34', None, None, "Missing"]
}

people_df = pd.DataFrame(people)
```

```python
# dropna method 

people_df.dropna()  # default arguments for dropna() are 

# dropna(axis = "index", how = "any")
# axis = "index" or "columns" means to drop row or columns 
# how = "any" means drop row/column when any of the value is NA
# how = "all" drops row/column when all values are missing
# to apply changes on original dataset use inplace = True

people_df.dropna(axis = "index", how = "all")

# Inorder to deal with "NA" and "Missing" we can use replace method on the dataframe 

people_df.replace("NA", np.nan, inplace = True)
people_df.replace("Missing", np.nan, inplace = True)
```

```python
# We are fine if first , last = Na, but we need to drop the row if email = Na 

people_df.dropna(axis = "index", how = "any", subset = ["email"])  # since we are passing the subset argument here , how is not doing much becuase only email column is checked
```

```python
# we need either last name or email for the daata to be useful 

people_df.dropna(axis = "index", how = "all", subset = ["last", "email"])
```

```python
# we can use isna() mask to see which values in tables will be treated as na 

people_df.isna()
```

```python
# fillna() to replace na values 

people_df.fillna("MISSING")
```

```python
# Inorder to get datatypes of all columns 

people_df.dtypes
```

```python
# Type Casting

# astyoe() method is used for type casting it can act on both Series and dataframe
# np.nan values are actually float values so while converting a column with a nan value to number convert it to float because converting 
# it to int will give error

people_df["age"] = people_df["age"].astype(float)
```
